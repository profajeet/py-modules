```python
#Import os Library
import os

# Fork a child process
processid = os.fork()
print(processid)

# processid > 0 represents the parent process
if processid > 0 :
  print("\nParent Process:")
  print("Process ID:", os.getpid())
  print("Child's process ID:", processid)

# processid = 0 represents the created child process
else :
  print("\nChild Process:")
  print("Process ID:", os.getpid())
  print("Parent's process ID:", os.getppid())
```

```python
import os

# Check access with os.F_OK
path1 = os.access("data.json", os.F_OK)
print("Does the path exists:", path1)

# Check access with os.R_OK
path2 = os.access("data.json", os.R_OK)
print("Access to read the file:", path2)

# Check access with os.W_OK
path3 = os.access("data.json", os.W_OK)
print("Access to write to file:", path3)

# Check access with os.X_OK
path4 = os.access("data.json", os.X_OK)
print("Can path be executed:", path4)
```
